# Generate Datasets

The next few cell sections, we will be generating datasets. We need the following datasets:

- First one is a log-normally distributed dataset. 
- The second one is a linear regression dataset.
- The third is a logistic regression dataset. 
- The final one is an exponential decay curve dataset.

For each dataset, there are two groups: one with a lot of data (100 data points), the other with substantially fewer data points (10 data points).

We will then create a panel of plots:

- The first column shows just the raw data.
- The second column shows a "point estimate".
- The third column shows a Bayesian estimate.
- The final column shows a hierarchical estimate for two groups.

This implies that the data generating process is hierarchical in nature.

## Generate Lognormal Datasets

```python
np.random.seed(42)
mu = 5
sd = 1
mu1 = np.random.normal(loc=mu, scale=sd)
mu2 = np.random.normal(loc=mu, scale=sd)
data1 = np.random.lognormal(mean=mu1, sigma=1.0, size=100)
data2 = np.random.lognormal(mean=mu2, sigma=1.0, size=10)

df1 = pd.DataFrame(data1)
df1['group'] = "0"

df2 = pd.DataFrame(data2)
df2['group'] = "1"
df = pd.concat([df1, df2]).reset_index(drop=True).rename(columns={0: 'measurement'})
```
