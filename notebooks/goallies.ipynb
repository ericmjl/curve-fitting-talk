```python
data = pd.read_csv('data/hockey-goalie.csv', skiprows=1)
df = data.groupby('Player').sum().reset_index()
df['SV%_adj'] = df['SV%'] / data.groupby('Player').size().values
df['Player'] = df['Player'].apply(lambda x: x.split('\\')[0])
df[['Player', 'SA', 'SV', 'SV%_adj']].head(20)
```

# No Pooling

```python
with pm.Model() as nopool:
    n_players = len(df)
    p = pm.Beta('p', alpha=1, beta=1, shape=(n_players,))
    
    like = pm.Binomial('likelihood', n=df['SA'], p=p, observed=df['SV'])
    
    trace = pm.sample(2000)
```

```python
pm.traceplot(trace)
```

```python
import seaborn as sns
```

```python
fig = plt.figure(figsize=(6, 18))
pm.forestplot(trace, varnames=['p'], ylabels=df['Player'])
```

```python
p_df = pd.DataFrame(trace['p'].T)
p_df = (p_df.join(df['Player'])
        .melt(id_vars='Player')
        .rename_column('variable', 'iter')
        .rename_column('value', 'p')
       )
p_df.to_csv('data/goalie_nopool_posterior.csv')
```

```python
ds = hv.Dataset(p_df, kdims=['Player', 'iter'], vdims=['p'])
players = ['Dylan Ferguson', 'Scott Foster', "Jake Allen"]
ds.select(Player=players).to(hv.Curve, kdims=['iter'], vdims=['p']).grid('Player')
```

```python
from ipywidgets import interact
```

```python
@interact(player=list(df['Player']))
def plot_player_performance(player):
    c = p_df['Player'] == player
    ax = sns.boxplot(data=p_df[c], x='Player', y='p')
    ax.set_ylim(0, 1)
    
    percentage = df.set_index('Player').loc[player, 'SV%']
    sa = df.set_index('Player').loc[player, 'SA']
    ax.set_title(f'SV%:{percentage:.2f}, N:{sa}')
    
    
```

```python
import theano.tensor as tt

with pm.Model() as pool:
    n_players = len(df)

    phi = pm.Uniform('phi', lower=0.0, upper=1.0)
    kappa_log = pm.Exponential('kappa_log', lam=1.5)
    kappa = pm.Deterministic('kappa', tt.exp(kappa_log))

    p = pm.Beta('p', alpha=phi*kappa, beta=(1.0-phi)*kappa, shape=(n_players,))
    
    like = pm.Binomial('likelihood', n=df['SA'], p=p, observed=df['SV'])
    
    trace_pool = pm.sample(2000)
```

```python
fig = plt.figure(figsize=(6, 18))
pm.forestplot(trace_pool, varnames=['p'], ylabels=df['Player'])
```

```python
@interact(player=list(df['Player']))
def plot_player_performance(player):
    c = p_df['Player'] == player
    ax = sns.boxplot(data=p_df[c], x='Player', y='p')
    ax.set_ylim(0, 1)
    
    percentage = df.set_index('Player').loc[player, 'SV%']
    sa = df.set_index('Player').loc[player, 'SA']
    ax.set_title(f'SV%:{percentage:.2f}, N:{sa}')
```

```python
p_dfpool = pd.DataFrame(trace_pool['p'].T)
p_dfpool = (p_dfpool.join(df['Player'])
            .melt(id_vars='Player')
            .rename_column('variable', 'iter')
            .rename_column('value', 'p')
           )
p_dfpool.to_csv('data/goalie_pool_posterior.csv')
```

```python
ds = hv.Dataset(p_dfpool, kdims=['Player'], vdims=['p'])
ds.select(Player=players).to(hv.BoxWhisker, kdims=['Player'], vdims=['p'])
```

```python
# Shrinkage plot.
shrinkage = pd.DataFrame()
shrinkage['No Pooling'] = p_df.groupby('Player').mean()['p']
shrinkage['Partial Pooling'] = p_dfpool.groupby('Player').mean()['p']
shrinkage['Full Pooling'] = [df['SV%_adj'].mean()] * len(df)
shrinkage = shrinkage.reset_index()
shrinkage = shrinkage.melt(id_vars='Player').rename_column('variable', 'Pooling').rename_column('value', 'p')

ds = hv.Dataset(shrinkage, vdims=['p'], kdims=['Pooling', 'Player'])
ds.to(hv.Curve).overlay('Player')
# shrinkage
```
