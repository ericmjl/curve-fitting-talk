```python
import imcmc
from pathlib import Path

img_dir = Path('../assets/images/')
```

```python
mit = imcmc.load_image(img_dir / 'mit-logo.png', 'L')
mit_trace = imcmc.sample_grayscale(mit, samples=1500, tune=500, nchains=20)
imcmc.plot_multitrace(mit_trace, mit, marker='o', markersize=10,
                      colors=['#973436'] * 4 + ['#666666'], alpha=0.9);
```

```python
imcmc.make_gif(mit_trace, mit, dpi=40, marker='o', markersize=10,
               colors=['#973436'] * 4 + ['#666666'], alpha=0.9,
               filename=img_dir / 'mit.gif')
```

```python
novartis = imcmc.load_image(img_dir / 'novartis-logo-modified.png', 'L')
novartis_trace = imcmc.sample_grayscale(novartis, samples=500, tune=500, nchains=12)
imcmc.plot_multitrace(novartis_trace, novartis, marker='o', markersize=10,
                      colors=['#0079bb'] * 7 + ['#f27c30'] * 3 + ['#eea82f'] * 2, alpha=0.9);
```

```python
from random import shuffle
colors = ['#0079bb'] * 7 + ['#f27c30'] * 3 + ['#eea82f'] * 2
shuffle(colors)
```

```python
imcmc.make_gif(novartis_trace, novartis, dpi=40, marker='o', markersize=10,
               colors=colors, alpha=0.9,
               filename=img_dir / 'novartis.gif')
```

```python
python = imcmc.load_image(img_dir / 'python-logo.png', 'L')
# python = python * python > 0
python_trace = imcmc.sample_grayscale(python, samples=1500, tune=500, nchains=10, threshold=0.1)
colors = ['#fed050'] * 5 + ['#3a73a1'] * 5
shuffle(colors)
imcmc.plot_multitrace(python_trace, python, marker='o', markersize=10,
                      colors=colors, alpha=0.9);
```

```python
imcmc.make_gif(python_trace, python, dpi=40, marker='o', markersize=10,
               colors=colors, alpha=0.9,
               filename=img_dir / 'python.gif')
```
